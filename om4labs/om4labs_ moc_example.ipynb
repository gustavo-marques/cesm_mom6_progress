### OM4-lab construction of a dictionary 

- outdir  = output dir

- static = static file

- infile = netcdf data to be read

- topog = topography file (static file)

- hgrid = supergrid

- interactive = True (if using notebook)

- format = None, png, etc

- label = name of the run
