```python
# load WOA18 data
obs_path = '/glade/u/home/gmarques/Notebooks/CESM_MOM6/WOA18_remapping/'
woa = xr.open_dataset(obs_path+'woa18_decav_merged_monthly_deep_04.nc', 
                             decode_times=True)
```

```python
woa
```

```python
months = ['JAN','FEB','MAR','JUL','AGU','SEP']
indices = [0,1,2,6,7,8]

for m, t in zip(months,indices):
    salt = woa.s_an[t,:]
    temp = woa.theta0[t,:]
    ds = temp.to_dataset(name='temp')
    ds1 = ds.assign(salt=salt).drop(['time'])
    fname = 'woa18_{}.nc'.format(m)
    ds1.to_netcdf(fname)
```
